```python
path='/sa.jpg'
```

```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread(path)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
objs=DeepFace.analyze(img_path=path,actions=['age'])
print('age is',objs[0]['age'])


```

```python
objs=DeepFace.analyze(img_path=path,actions=['emotion'])
print('emotion is',objs[0]['dominant_emotion'])
```

```python
objs=DeepFace.analyze(img_path=path,actions=['gender'])
print('gender is',objs[0]['dominant_gender'])
```

```python
objs=DeepFace.analyze(img_path=path,actions=['race'])
print('race is',objs[0]['dominant_race'])
```
